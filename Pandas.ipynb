```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sample_data = {
    'OrderDate': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
    'Total': [100, 120, 90, 150, 110]
}
sample_df = pd.DataFrame(sample_data)

display(sample_df)
plt.figure(figsize = (12,6))
plt.plot(sample_df['OrderDate'], sample_df['Total'])
plt.title('Sample Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
